```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st


# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
data = pd.merge(study_results, mouse_metadata, how= "left", on = [ "Mouse ID", "Mouse ID"])
# Display the data table for preview
data.head()
```

```python
# Checking the number of mice.
mice= data["Mouse ID"].unique()
mice_count = len (mice)
mice_count
print('Total number of mice =  ' + str(mice_count))
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 

#duplicateRows = df[df.duplicated()]
duplicate_mice = data [data.duplicated(subset=['Mouse ID','Timepoint'])]["Mouse ID"].unique()
duplicate_mice
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
dublicate_mice_data = data[data["Mouse ID"] == duplicate_mice[0]]
dublicate_mice_data
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_data = data[data["Mouse ID"] != duplicate_mice[0]]
clean_data. head()
```

```python
# Checking the number of mice in the clean DataFrame.
clean_count=len(clean_data["Mouse ID"].unique())
clean_count
print('Total number of mice (no duplicate) =  ' + str(clean_count))
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
drug_mean = clean_data.groupby(["Drug Regimen"]).mean()["Tumor Volume (mm3)"]
drug_median = clean_data.groupby(["Drug Regimen"]).median()["Tumor Volume (mm3)"]
drug_var = clean_data.groupby(["Drug Regimen"]).var()["Tumor Volume (mm3)"]
drug_std = clean_data.groupby(["Drug Regimen"]).std()["Tumor Volume (mm3)"]
drug_sem = clean_data.groupby(["Drug Regimen"]).sem()["Tumor Volume (mm3)"]

# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.
drug_stat_summary = pd.DataFrame({"mean": drug_mean,"median": drug_median,"var": drug_var,"std": drug_std,"sem": drug_sem})
drug_stat_summary
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, 
# and SEM of the tumor volume for each regimen
# Using the aggregation method, produce the same summary statistics in a single line.

drug_stat_summary_aggregate  = clean_data.groupby(["Drug Regimen"]).aggregate(['mean', 'median', 'var', 'std', 'sem'])["Tumor Volume (mm3)"]
drug_stat_summary_aggregate
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
t= clean_data.groupby("Mouse ID").max()["Timepoint"]
t
# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
greatest_timepoint = pd.DataFrame(t)

greatest_timepoint.rename(columns = {'Timepoint':'Greatest_timepoint'}, inplace = True)
greatest_timepoint

clean_data_max_timepoint = pd.merge(clean_data,greatest_timepoint, how="left", on=["Mouse ID","Mouse ID"])
clean_data_max_timepoint
clean_data_max_timepoint["tumor_vol@last_time"] = ""
clean_data_max_timepoint

for j in range (0,clean_data_max_timepoint.shape[0]):
    for i in range (0,clean_data_max_timepoint.shape[0]):
        if clean_data_max_timepoint.iloc[i,0] ==  clean_data_max_timepoint["Mouse ID"][j]:
            if clean_data_max_timepoint.iloc[i,1] ==  clean_data_max_timepoint["Greatest_timepoint"][j]:
                clean_data_max_timepoint["tumor_vol@last_time"][j] = clean_data_max_timepoint.iloc[i,2]
clean_data_max_timepoint
#clean_data_max_timepoint.to_csv('test.csv')
```

```python
# Put treatments into a list for for loop (and later for plot labels)

# Create empty list to fill with tumor vol data (for plotting)

Capomulin_outlier=[]
Ramicane_outlier=[]
Infubinol_outlier=[]
Ceftamin_outlier=[]

# Calculate the IQR and quantitatively determine if there are any potential outliers. 

    # Locate the rows which contain mice on each drug and get the tumor volumes
    
    
    # add subset 
    
    
    # Determine outliers using upper and lower bounds
    
#____________________Capomulin

t = clean_data_max_timepoint.loc[clean_data_max_timepoint['Drug Regimen'] == "Capomulin"]
#t
t1 = t["tumor_vol@last_time"]
#t1

quartiles = t1.quantile([.25,.5,.75])
lowerq = quartiles[0.25]
upperq = quartiles[0.75]
iqr = upperq-lowerq

lower_bound = lowerq - (1.5*iqr)
upper_bound = upperq + (1.5*iqr)

for j in t1:
    if (j < lower_bound) | (j > upper_bound):
        if j not in Capomulin_outlier:
            Capomulin_outlier.append(j)
print("Capomulin's potential outliers: " + str(Ceftamin_outlier))

#_____________________Ramicane

m = clean_data_max_timepoint.loc[clean_data_max_timepoint['Drug Regimen'] == "Ramicane"]
#t
m1 = m["tumor_vol@last_time"]
#t1

quartiles = m1.quantile([.25,.5,.75])
lowerq = quartiles[0.25]
upperq = quartiles[0.75]
iqr = upperq-lowerq

lower_bound = lowerq - (1.5*iqr)
upper_bound = upperq + (1.5*iqr)

for j in m1:
    if (j < lower_bound) | (j > upper_bound):
        if j not in Ramicane_outlier:
            Ramicane_outlier.append(j)

print("Ramicane's potential outliers: " + str(Ceftamin_outlier))

#___________________Infubinol

l = clean_data_max_timepoint.loc[clean_data_max_timepoint['Drug Regimen'] == "Infubinol"]
#t
l1 = l["tumor_vol@last_time"]
#t1

quartiles = l1.quantile([.25,.5,.75])
lowerq = quartiles[0.25]
upperq = quartiles[0.75]
iqr = upperq-lowerq

lower_bound = lowerq - (1.5*iqr)
upper_bound = upperq + (1.5*iqr)

for j in l1:
    if (j < lower_bound) | (j > upper_bound):
        if j not in Infubinol_outlier:
            Infubinol_outlier.append(j)
            
print("Infubinol's potential outliers: " + str(Infubinol_outlier))

#________________________Ceftamin

w = clean_data_max_timepoint.loc[clean_data_max_timepoint['Drug Regimen'] == "Ceftamin"]
#t
w1 = w["tumor_vol@last_time"]
#t1

quartiles = w1.quantile([.25,.5,.75])
lowerq = quartiles[0.25]
upperq = quartiles[0.75]
iqr = upperq-lowerq

lower_bound = lowerq - (1.5*iqr)
upper_bound = upperq + (1.5*iqr)

for j in w1:
    if (j < lower_bound) | (j > upper_bound):
        if j not in Ceftamin_outlier:
            Ceftamin_outlier.append(j)
        
print("Ceftamin's potential outliers: " + str(Ceftamin_outlier))

```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
from scipy.stats import linregress
x_values = k2['Weight (g)']
y_values = k2['Tumor Volume (mm3)']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(5.8,0.8),fontsize=15,color="red")
plt.xlabel('weight(g)')
plt.ylabel('Average Tumor Volume (mm3)')

correlation = st.pearsonr(x_values,y_values)
print(f"The correlation between mouse weight and the average tumor volume {round(correlation[0],2)}")

plt.show()

```
